```python
import pandas as pd
import numpy as np
```

# Data cleaning - literature dataset
## Import dataset

```python
# load given dataset
lit_df = pd.read_csv('Literature-data_TU-Darmstadt.txt', sep = '\t', header = 1, skiprows = [3])
lit_df
```

## Keep needed attributes

```python
lit_clean = lit_df[['Autor', 'Titel', 'Jahr', 'Journal', 'Typ', 'DOI', 'Ausschlusspunkt']]
lit_clean
```

## Convert some string columns to lowercase

```python
lowercase_attr = ['Autor', 'Titel', 'Journal']
for i in lowercase_attr:
    lit_clean.loc[:, i] = lit_clean.loc[:, i].str.lower()
    
lit_clean
```

## Remove duplicates
### Rows with duplicate titles

```python
lit_clean[lit_clean['Titel'].duplicated()]
```

We find that row 782 and 800 are duplicates, but the DOI is missing in row 782, so we decide to keep row 800

```python
# 782 and 800 are duplicates
lit_clean[lit_clean['Titel'] == lit_clean.iloc[800]['Titel']]
```

```python
lit_clean.at[800, 'Ausschlusspunkt'] = lit_clean.at[782, 'Ausschlusspunkt']
```

```python
# 942 and 1012 are duplicates
lit_clean[lit_clean['Titel'] == lit_clean.iloc[1012]['Titel']]
```

```python
# 860 and 1059 are duplicates
lit_clean[lit_clean['Titel'] == lit_clean.iloc[1059]['Titel']]
```

```python
# remove rows 782, 1012, 1059
lit_clean = lit_clean.drop([782, 1012, 1059]).reset_index(drop = True)
```

### Rows with duplicate DOIs (not NaN or 0)

```python
x = lit_clean[lit_clean['DOI'].duplicated()]
x = x[(x['DOI'] != '0') & (x['DOI'].notnull())]
x
```

```python
# remove rows with duplicate DOIs
to_drop = list(x.index.values) # a list like so: [667, 668, ..., 1054, 1055]
lit_clean = lit_clean.drop(to_drop).reset_index(drop = True)
```

### Remaining rows with Ausschlusspunkt = Dopplung

```python
lit_clean[lit_clean['Ausschlusspunkt'] == 'Dopplung']
```

For the remaining 4 rows marked as 'Dopplung', duplicates couldn't be found for rows 666 and 973, so we keep these two rows.

```python
lit_clean = lit_clean.drop([681, 970]).reset_index(drop = True)
```

## Other modifications
We found rows with DOIs of 0 and NaN. Since our project idea is about how useful an article is based on the impact factor of its publishing journal, an invalid DOI is less relevant, so we decided to convert all null values to '0' for consistency instead of deleting those rows.

```python
print(str(len(lit_clean[lit_clean['DOI'] == '0'])) + ' rows with DOI = 0')
print(str(len(lit_clean[lit_clean['DOI'].isnull()])) + ' rows with DOI = NaN')
```

```python
lit_clean.loc[lit_clean['DOI'].isnull(), 'DOI'] = '0'
```

One of the values in the Typ column is 'Artikle'. We changed it to 'Article'.

```python
lit_clean[lit_clean['Typ'] == 'Artikle']
```

```python
lit_clean.loc[lit_clean['Typ'] == 'Artikle', 'Typ'] = 'Article'
```

## Export as csv file

```python
lit_clean
```

```python
wanted_attr = ['Titel', 'Autor', 'Jahr', 'Journal', 'Typ', 'DOI']
lit_clean[wanted_attr].to_csv('literature_dataset.csv', index = False)
```

# Data cleaning - impact factor dataset
## Import dataset

```python
# load dataset of journal impact factors
# from https://www.kaggle.com/datasets/mostafafaramin/scientific-journals-ranking-sjr
journal_df = pd.read_csv('journal_impact_factors_2020.csv')
journal_df
```

## Keep needed attributes

```python
rename_attr = {
    'Full Journal Title': 'Title',
    'Rank': 'Rank',
    'Journal Impact Factor': 'Impact_Factor',
    '5-Year Impact Factor': '5_Yr_IF',
    'Cited Half-Life': 'Half_Life',
    '% Articles in Citable Items': 'Percentage_Citable',
    'Average Journal Impact Factor Percentile': 'Avg_IF_Percentile'
}
journal_clean = journal_df.rename(columns = rename_attr)[list(rename_attr.values())]
journal_clean
```

## Drop rows with invalid impact factors
18 journals have no impact factor available. Since impact factor is the main property we will look at, we removed these rows

```python
to_drop = list(journal_clean[journal_clean['Impact_Factor'] == 'Not Available'].index.values)
journal_clean = journal_clean.drop(to_drop).reset_index(drop = True)
```

## Convert types and make title lowercase
- Most columns with floats as values have the type object. We replaced 'Not Available' values with NaN and converted these columns to floats
- Make the title column lowercase for consistency

```python
journal_clean.dtypes
```

```python
# convert the impact factor column
journal_clean['Impact_Factor'] = journal_clean['Impact_Factor'].astype(float)
```

```python
# convert other columns (that have 'Not Available' values) to float
def f(x):
    try:
        return np.float(x)
    except:
        return np.nan

convert_attr = ['5_Yr_IF', 'Half_Life', 'Percentage_Citable', 'Avg_IF_Percentile']
for i in convert_attr:
    journal_clean[i] = journal_clean[i].apply(f)
```

```python
# convert title column to lowercase
journal_clean.loc[:, 'Title'] = journal_clean.loc[:, 'Title'].str.lower()
```

## Export as csv file

```python
journal_clean
```

```python
journal_clean.to_csv('journal_dataset.csv', index = False)
```
